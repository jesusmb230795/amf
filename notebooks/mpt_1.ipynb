## Skewness


![title](Skewness_Feature.jpg)


Recall that the skewness is given by:

$$ S(R) = \frac{E[ (R-E(R))^3 ]}{\sigma_R^3} $$



```python
def skewness(r):
    demeaned_r = r - r.mean()
    # use the population standard deviation, so set dof=0
    sigma_r = r.std(ddof=0)
    exp = (demeaned_r**3).mean()
    return exp/sigma_r**3
```

```python
skewness(df_log).sort_values()
```

```python
normal_rets = np.random.normal(0, .2, (10000, 1))
normal_rets.mean(), normal_rets.std()
```

```python
skewness(normal_rets)
```

# Kurtosis

Intuitively, the kurtosis measures the "fatness" of the tails of the distribution. The normal distribution has a kurtosis of 3 and so if the kurtosis of your returns is less than 3 then it tends to have thinner tails, and if the kurtosis is greater than 3 then the distribution has fatter tails.

Kurtosis is given by:

$$ K(R) = \frac{E[ (R-E(R))^4 ]}{\sigma_R^4} $$

This is very similar to the skewness, so we can just copy and paste it and then edit it to compute the 4th rather than the 3rd power (as was the case for skewness).

```python
def kurtosis(r):
    demeaned_r = r - r.mean()
    # use the population standard deviation, so set dof=0
    sigma_r = r.std(ddof=0)
    exp = (demeaned_r**4).mean()
    return exp/sigma_r**4
```

```python
kurtosis(df_log)
```

```python
kurtosis(normal_rets)
```

```python
ex1_r = np.array([[2.07],
                 [0.21],
                 [1.01]])
ex1_cov = np.array([[48.2, 7.82, -2.65],
                    [7.82, 16.34, 0.99 ],
                    [-2.65, 0.99, 34.25]
                   ])

ex1_1 = np.array([[1],
                  [1],
                  [1]])
```

```python
def wp_op_from_rp(E, Cov, rp, M_1):
    n = E.shape[0]
#     M_1 = (np.zeros(n) + 1 )
    A = (np.matmul(M_1.T, inv(Cov)) @ E).item() 
    B = (E.T @ inv(Cov) @ E).item()
    C = (M_1.T @ inv(Cov) @ M_1).item()
    D = B*C - A**2
    g = (1/D) * (B*inv(Cov) @ M_1 - A * inv(Cov) @ E)
    h = (1/D) * (C * inv(Cov) @ E - A * inv(Cov) @ M_1)
    
    W = g + h * rp
    
    rp_c = (W.T @ ex1_r).item()
    var_p = round(((W.T @ Cov @ W)).item(),4)
    std_p = round(((W.T @ Cov @ W)).item()**(1/2), 4)
    print("El rendimiento del portafolio: {}".format(rp_c))
    print("La varianza del portafolio: {}".format(var_p))
    print("El riesgo del portafolio: {}".format(std_p))
    
    return W
    
```

```python
def wp_op_from_rp_with_rf(E, Cov, rp, M_1, rf):

    A = (np.matmul(M_1.T, inv(Cov)) @ E).item() 
    B = (E.T @ inv(Cov) @ E).item()
    C = (M_1.T @ inv(Cov) @ M_1).item()
    
    H = B -2*rf*A+rf**2*C
    
    Wp = inv(Cov) @ (E - (rf * M_1)) * ((rp - rf)/H)

    Wf = 1 - Wp.T @ M_1
    
    rp_c = (Wp.T @ ex1_r).item() + Wf.item()*rf
    var_p = round(((Wp.T @ Cov @ Wp)).item(),4)
    std_p = round(((Wp.T @ Cov @ Wp)).item()**(1/2), 4)
    print("El rendimiento del portafolio: {}".format(rp_c))
    print("La varianza del portafolio: {}".format(var_p))
    print("El riesgo del portafolio: {}".format(std_p))
    
    return Wp, Wf

wp_op_from_rp_with_rf(ex1_r, ex1_cov, 1.5, ex1_1, 0.18)
```
